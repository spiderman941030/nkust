```python
def determine_quadrant(x, y):
    if x == 0 and y == 0:
        return "原點"
    elif x == 0:
        return "y軸"
    elif y == 0:
        return "x軸"
    elif x > 0 and y > 0:
        return "第一象限"
    elif x < 0 and y > 0:
        return "第二象限"
    elif x < 0 and y < 0:
        return "第三象限"
    elif x > 0 and y < 0:
        return "第四象限"
input_str = input("請輸入座標 (x, y): ")
input_str = input_str.replace(" ", "").replace("(", "").replace(")", "")
coordinates = input_str.split(",")
if len(coordinates) >= 2:
    x = float(coordinates[0])
    y = float(coordinates[1])
else:
    x = float(input("請輸入 x 座標: "))
    y = float(input("請輸入 y 座標: "))
result = determine_quadrant(x, y)
print(f"該點位於: {result}")
```

```python
# 台灣縣市和今日溫度的對應
cities = ["台北", "新北", "桃園", "台中", "台南", "高雄", "宜蘭", "花蓮", "嘉義", "屏東"]
temperatures = [25, 27, 24, 22, 28, 30, 23, 21, 26, 29]  # 假設的今日溫度

# 使用者輸入
user_input = input("請輸入台灣縣市: ")

# 查找該縣市的溫度
if user_input in cities:
    index = cities.index(user_input)
    temperature = temperatures[index]
    print(f"{user_input} 今天的溫度是 {temperature}°C")
else:
    print("未找到該縣市，請檢查輸入是否正確。")
```
